```python
# Task 2: Data Cleaning and Preparation

## Objective
The objective of this task is to clean and prepare the Global Superstore dataset for further data analysis.

## Dataset
The dataset contains sales transaction data including customer details, product information, order dates, and financial metrics such as sales, profit, and shipping cost.

## Tools Used
- Python
- Pandas
- Google Colab

## Data Cleaning Steps
1. Loaded the dataset using Pandas
2. Examined dataset structure and column types
3. Identified and handled missing values
4. Removed duplicate rows
5. Converted date columns to datetime format
6. Verified data consistency
7. Exported the cleaned dataset as a CSV file
```

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Global_Superstore2.csv", encoding='latin1')
```

```python
df.to_csv("cleaned_superstore.csv", index=False)
```
